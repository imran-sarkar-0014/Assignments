# Question 1.(8 points)
In this task, you have to work with a dataset in the file students.csv. The data provides
some information about students, including: (i) gender, (ii) the educational background of the student’s
parents, (iii) whether the student has participated in test preparation courses, and (iv) the test scores
in math, reading and writing. By making use of the basic data analysis tools introduced in the course,
answer the following questions:
• Are there more male or female students included in the dataset?
• In which subject did the students perform best overall (i.e. what is the subject with the highest
average score)?
• How many students can you find with perfect results (100 score in all three subjects)?
• Can we say that participating in the test preparation course improves the results in math (i.e. is the
average math score higher for those who completed the test preparation course)?

# Answer to the question no 1

```python
#read the students.csv file and store it's dataframe into students variable
students = pd.read_csv('students.csv')

#show students dataframe's data
students
```

question 1.1: Are there more male or female students included in the dataset?

```python
#count the repetition of gender's value and store in gender as hash table.
genders = students['gender'].value_counts()

# compare between total male and total female value and
# print the maximum 
if genders['male'] < genders['female']:
    print('Female is more.')
else:
    print('Male is more.')
```

question 1.2: In which subject did the students perform best overall (i.e. what is the subject with the highest
average score)?

```python
# calculate the average score of math and store in avg_math_score
avg_math_score = students['math_score'].mean()

# calculate the average score of reading and store in avg_reading_score
avg_reading_score = students['reading_score'].mean()

# calculate the average score of writing and store in avg_writing_score
avg_writing_score = students['writing_score'].mean()


# store all subject average in a table named scores and 
# avgerage value is the key and subject name is the value
scores = {
     avg_math_score : 'math score',
     avg_reading_score : 'reading score',
     avg_writing_score: "writing score"
}


# find the maximum key from score and store in max_score_subject
max_score_subject = max(scores)

# find the value of max_score_subject's key to get the highest average subject name.
max_score_value = scores[max_score_subject]

# print the results 
print(f'{max_score_value} with the highest average score -> {max_score_subject}')
```

question 1.3: How many students can you find with perfect results (100 score in all three subjects)?


```python
# find the students who get full marks (100) score in all three subjects and 
# store in topper.
topper =  students.loc[
    (students["math_score"] == 100) & 
    (students['reading_score'] == 100) & 
    ( students['writing_score'] == 100 )
]

# convert dataframe to numpy array for count the total row of topper
toppers_row = topper.to_numpy()

# get the shape of toppers_row and it's first value is the total row number.
(total_row,row) = toppers_row.shape

print(f'{total_row} students with perfect results (100 score).')
```

question 1.4: Can we say that participating in the test preparation course improves the results in math (i.e. is the
average math score higher for those who completed the test preparation course)?

```python
# get a dataframe's of students who completed their test preparation course and
# store in df_test_course_complated
df_test_course_complated = students.loc[students["test_prep_course"] == "completed"]

# calculate the average of df_test_course_complated and
# store in test_course_complated_avarage
test_course_complated_avarage = df_test_course_complated["math_score"].mean()

# get a dataframe's of students who didn't completed their test preparation course and
# store in df_test_course_none
df_test_course_none = students.loc[students["test_prep_course"] == "none"]

# calculate the average of df_test_course_none and
# store in test_course_none_avarage
test_course_none_avarage = df_test_course_none["math_score"].mean()

# conditionally print the result
if(test_course_complated_avarage > test_course_none_avarage):
    print('Yes, participating in the test preparation course improves the results in math.')
else:
    print('No, participating in the test preparation course not improves the results in math')
```

# Question 2.(7 points)
In this exercise, you will have to analyze a dataset about some employees of a company
(salary.csv). The data includes information about (i) years of experience of the employee, (ii) the age of
the employee, and (iii) the salary of the employee. You need to write the code to answer the following
questions to transform and understand the data further.


```python
# read the salary.csv file and store in salary variable
salary = pd.read_csv('salary.csv')
salary
```

question 2.1: There are some missing values in the data. In the experience years column, replace all missing values
by 3, and in the age column replace all the missing values by the mean of the column.


```python
# calculate the average of age and store in avg_age
avg_age = int(salary["age"].mean())

# replace missing values
salary = salary.fillna({
    # In the experience years column, replace all missing values by 3
    'experience_years': 3,
    
    # In the age column replace all the missing values by the mean of the column.
    'age' : avg_age,
})
salary
```

# Question 2.2
Check the correlation between the columns of the data. Based on the results, what is your opinion,
does experience or age impact salary more?


```python
# calculate the correlation between salary and experience_years 
corr_experience =salary['salary'].corr(salary['experience_years'])

# calculate the correlation between salary and age 
corr_age = salary['salary'].corr(salary['age'])

# 
if(corr_age > corr_experience):
    print('Age impact more.')
else:
    print('Experience impact more.')
```

question 2.3: Create a categorical version of the salary column with 3 groups: (i) values between 0 and 55000, (ii)
values between 55000 and 90000, (iii) values between 90000 and 125000. Label the groups [’Low’,
’Medium’, ’High’]. Calculate the average age of employees who have High salary.

```python

# the catagorical_salary table stores the dataframes of three groups with it's name.
catagorical_salary = {

     # get all employees with the salary values between 0 and 55000
    'Low' : salary.loc[(salary['salary'] >= 0) & (salary['salary'] < 55000) ],
    
    # get all employees with the salary values between 55000 and 90000
    'Medium': salary.loc[(salary['salary'] >= 55000) & (salary['salary'] < 90000) ],
    
    # get all employees with the salary values between 90000 and 125000
    'High' : salary.loc[(salary['salary'] >= 90000) & (salary['salary'] <= 125000) ]
}

#Calculate the average age of employees who have High salary and
# store in avarage_age_of_high_salary casting into integer.
avarage_age_of_high_salary = int(catagorical_salary['High']['age'].mean())

# print the final result
print('avarage ege of high salary is :', avarage_age_of_high_salary)
```

# 3. (10 points) 
In this exercise, you will have to work with a dataset about movies (movies.csv). In the
dataset, you can find information about: (i) movie title, (ii) short overview of the movie, (iii) the original
language, (iv) number of votes given by people who have seen the movie, and (v) the average of the votes

```python
# read the movies.csv and store in movies
movies = pd.read_csv('movies.csv')
movies
```

Question 3.1: Evaluations that are based on too few votes cannot be completely trusted. For this reason, start
by removing movies that has less than 100 votes (vote count). Perform the following tasks on this
filtered dataset.

```python
# removed all movies which has less than 100 vote and
# stored in filterd_movies variable

filterd_movies = movies.loc[movies['vote_count'] >= 100]

# show the filterd_movies
filterd_movies
```

Question 4.5 : Create a categorical version of the vote average column with 4 groups: (i) values between 0 and 2.5,
(ii) values between 2.5 and 5, (iii) values between 5 and 7.5, and (iv) values between 7.5 and 10.
Label the groups [’Bad’, ’OK’, ’Good’, ’Excellent’]. Which of the four created groups has the highest
number of movies?

```python

# catagorical_movies is a table of movies with rating key.
catagorical_movies = {
    
    #values between 0 and 2.5,
    'Bad' : movies.loc[movies['vote_average'] <= 2.5] ,
    
    #values between 2.5 and 5
    'OK' : movies.loc[(movies['vote_average'] >= 2.5) & (movies['vote_average'] < 5) ],
    
    #values between 5 and 7.5,
    'Good': movies.loc[(movies['vote_average'] >= 5) & (movies['vote_average'] < 7.5) ],
    
    #values between 7.5 and 10.
    'Excellent' : movies.loc[(movies['vote_average'] >= 7.5) & (movies['vote_average'] <= 10) ]
}

# this variable is to keep tracking the hightest number of movies with it's catagorical name.
number_of_movies = {
    'name': '',
    'count': 0
}



# iterate over the catagorical_movies
for c in catagorical_movies:
    
    # check if  catagorical_movies's current size if grater then older
    if( catagorical_movies[c].size > number_of_movies['count']):
        
        # if current size is older then store it
        number_of_movies = {
            'name':c,
            'count': catagorical_movies[c].size
        }

# print the final result
print(f"The '{number_of_movies['name']}' group has the hightest number of movies.")
```
